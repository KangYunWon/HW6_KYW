```python
import csv
import matplotlib.pyplot as plt
import numpy as np

def main():
    plt.rcParams['font.family'] ='Malgun Gothic'
    plt.rcParams['axes.unicode_minus'] =False #한국어 깨짐 방지용 코드
    
    years=['2002', '2007', '2012', '2017', '2022'] #인구수 조사를 시행한 연도를 저장할 변수
    male = [] #연도별 남성 인구수를 저장할 배열
    female = [] #연도별 여성 인구수를 저장할 배열
    total = [] #연도별 총 인구수를 저장할 배열
    mcnt = 0 #남성 인구수가 많은 횟수 저장용 변수
    scnt = 0 #남성 인구수와 여성 인구수가 같은 횟수 저장용 변수
    #여성 인구수가 만흥 횟수는 mcnt와 scnt를 이용해 구함
    
    f = open('jeju population.csv', 'r', encoding='ansi')
    data = csv.reader(f)
    header = next(data)
    for row in data:
        for i in range(2,5):
            row[i]=int(row[i])
            if i == 2:
                total.append(row[i])
            elif i == 3:
                male.append(row[i])
            else:
                female.append(row[i])
            
    f.close()
            
    bar_width = 0.4
    bar_positions_male = np.arange(len(years)) #남성 인구수를 표시하는 막대의 위치를 x축의 연도에 맞게 배치
    bar_positions_female = bar_positions_male + bar_width #여성 인구수를 표시하는 막대는 남성 막대 바로 옆에 배치
    fig, ax = plt.subplots() #그래프를 배치하기 위해 subplot 생성
    ax.bar(bar_positions_male, male, bar_width, label='남성', edgecolor='white', color = '#1d9bfc')
    ax.bar(bar_positions_female, female, bar_width, label='여성', edgecolor='white', color = '#ed563b')
    #막대 위치 지정, 자료지정, 막대 가로 길이 지정, 범례 추가, 테두리 색 지정, 막대 색 지정
    ax.set_xlabel('연도') #x축은 인구 조사를 시행한 연도를 의미
    ax.set_ylabel('인구수(명)') #y축은 인구수를 의미
    ax.set_title('제주도의 20년간 남녀 인구수 변화 추이(2002년 ~ 2022년)')
    ax.set_xticks(bar_positions_male + bar_width / 2) #x축 눈금의 간격 조절
    ax.set_xticklabels(years) #x축 눈금에 들어갈 내용지정
    ax.legend() #범례 보이기
    plt.show() #그래프 보이기
    
    #결과 출력
    for i in range(len(years)): #인구수 조사를 시행한 모든 연도에 대해서 실행
        if male[i] > female[i]:
            result = "남성의 인구수가 여성의 인구수보다 {0}명 많다.".format(male[i]-female[i])
            mcnt += 1
        elif female[i] > male[i]:
            result = "여성의 인구수가 남성의 인구수보다 {0}명 많다.".format(female[i]-male[i])
        else:
            result = "여성의 인구수와 남성의 인구수가 같다."
            scnt += 1
        print("{0}년의 총 인구수는 {1}명, 여성 인구수는 {2}명, 남성 인구수는 {3}명으로 {4}"
              .format(years[i], total[i], female[i], male[i], result))
    
    print((0))
    
if __name__ == "__main__":
    main()
```
